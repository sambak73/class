# Load the file

```python
with open('miracle_in_the_andes.txt', 'r') as file:
    book = file.read()
```

# Identify # of chapters

### using string method

```python
print(book.count('Chapter'))
```

### using RegExp

```python
import re
```

```python
pattern = re.compile('Chapter [0-9]+')
```

```python
re.findall(pattern, book)
findings = re.findall(pattern, book)
len(findings)
```

## list of sentences that contain word love

```python
patternl = re.compile('[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.')
findings = re.findall(patternl, book)
findings
```

```python
d = {}
patternw = re.compile('[a-zA-z]+')
findings = re.findall(patternw, book.lower())
findings
len(findings)
for word in findings:
    if word in d:
        d[word] = d[word] + 1
    else:
        d[word] = 1
d_list = [(value, key) for (key, value) in d.items()]
```

```python
patternp = re.compile('[^\n]+[^a-zA-Z]+love+[^a-zA-Z]+[^\n]*.')
findings = re.findall(patternp, book)
findings[:2]
```

```python
patterntitle = re.compile('([a-zA-Z ]+)\n\n')
findings = re.findall(patterntitle, book)
findings
```

```python
findings = re.findall(re.compile('[a-zA-Z ]+'), book)
findings              
```
